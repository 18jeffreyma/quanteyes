```python
import os
import re

import numpy as np
from PIL import Image
import torch
```

```python
path = '/Users/ericdong/Desktop/Harvard/FA23/CS 249R/quanteyes/data_bak/2bit-octree'
files = os.listdir(path)
files = sorted(files)
```

```python
for file in files:
	path_file = os.path.join(path, file)
	img = Image.open(path_file)
	img = np.expand_dims(np.array(img), axis=0)
	img = torch.tensor(img)
	img = torch.nn.functional.max_pool2d(img.float(), 2).numpy()[0]
	img = Image.fromarray(img)
	if img.mode != 'L':
		img = img.convert('L')
	print(img.size)
	print(np.array(img).shape)
	# print(path_file)
	img.save(os.path.join(path, file))
```
