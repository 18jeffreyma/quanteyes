```python
from PIL import Image, ImageFilter
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
import cv2
from skimage import filters
```

```python
def remap_image(img):
	img = img - img.min()
	return (img / np.abs(img).max() * 255).astype(int)
```

```python
img = cv2.imread('/mnt/sdb/data/Openedsdata2020/openEDS2020-GazePrediction/train/sequences/6426/000.png')

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Blur the image for better edge detection
img = cv2.GaussianBlur(img, (31,31), 0)

plt.figure()
plt.imshow(remap_image(img), cmap='gray')

edges = cv2.Canny(image=img, threshold1=5, threshold2=15) # Canny Edge Detection
print(np.unique(edges))

plt.figure()
plt.imshow(remap_image(edges), cmap='gray')
```
