```python
import os
import torch
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from collections import Counter
import scipy.signal as signal
```

```python
data_dir = '/mnt/sdb/data/Openedsdata2020/openEDS2020-GazePrediction/train/sequences/' #6873/000.png'
directories = os.listdir(data_dir)
```

```python
count = 1
tops = []
bottoms = []
both = []
# max = 15
for directory in directories:
	files = os.listdir(data_dir + directory)
	img_path = data_dir + directory + '/' + np.random.choice(files)
	# print(img_path)
	img = plt.imread(img_path)
	# print(img.shape)
	top = img[30:130,220:420]
	bottom = img[270:370:,220:420]	
	# f, axarr = plt.subplots(1,3)
	# axarr[0].imshow(top,cmap='gray')
	# axarr[1].imshow(bottom,cmap='gray')
	# axarr[2].imshow(img,cmap='gray')
	# plt.show() 
	# print(img.size)
	# print("top: " + str(np.average(top)))
	# print("bottom: " + str(np.average(bottom)))
	tops.append(top)
	bottoms.append(bottom)
	both.append(top)
	both.append(bottom)
	# if count == max:
	# 	break
	# else:
	# 	count += 1

f, axarr = plt.subplots(1,3)
# f.set_figheight(30)
f.set_figwidth(30)
counts_top, bins_top = np.histogram(tops)
counts_bottom, bins_bottom = np.histogram(bottoms)
counts_both, bins_both = np.histogram(both)
axarr[0].stairs(counts_top, bins_top)
axarr[0].set_xlabel("top eye")
axarr[1].stairs(counts_bottom, bins_bottom)
axarr[1].set_xlabel("bottom eye")
axarr[2].stairs(counts_both, bins_both)
axarr[2].set_xlabel("top & bottom eye")
plt.show() 
```
